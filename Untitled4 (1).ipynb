```python
type_new = pd.get_dummies(data['type'], drop_first=True)
data_new = pd.concat([data, type_new], axis=1)
data_new.head()
```

```python
X = data_new.drop(['isFraud', 'type', 'nameOrig', 'nameDest'], axis=1)
y = data_new['isFraud']
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
	X, y, test_size=0.3, random_state=42)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score as ras
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
# Importing required libraries
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, accuracy_score  # Import metrics

# Ensure that you have the training and test data defined
# X_train, y_train, X_test, y_test should be defined before running this loop

# List of models to train
models = [
    LogisticRegression(),
    XGBClassifier(),
    SVC(kernel='rbf', probability=True),  # Make sure probability=True for SVC to get predict_proba
    RandomForestClassifier(n_estimators=7, criterion='entropy', random_state=7)
]

# Loop to train models and evaluate performance
for model in models:
    try:
        # Fitting the model to the training data
        model.fit(X_train, y_train)

        print(f'{model} : ')

        # Predicting probabilities for the training set (class 1)
        train_preds = model.predict_proba(X_train)[:, 1]

        # Calculate ROC AUC for training data
        train_auc = roc_auc_score(y_train, train_preds)

        # Calculate accuracy for training data
        train_accuracy = accuracy_score(y_train, model.predict(X_train))

        print(f'Training ROC AUC Score: {train_auc}')
        print(f'Training Accuracy: {train_accuracy}')

        # Predicting probabilities for the test set (class 1)
        y_preds = model.predict_proba(X_test)[:, 1]

        # Calculate ROC AUC for test data
        val_auc = roc_auc_score(y_test, y_preds)

        # Calculate accuracy for test data
        val_accuracy = accuracy_score(y_test, model.predict(X_test))

        print(f'Validation ROC AUC Score: {val_auc}')
        print(f'Validation Accuracy: {val_accuracy}')

        print()  # For better readability between models

    except Exception as e:
        print(f"Error with {model}: {e}")
```
